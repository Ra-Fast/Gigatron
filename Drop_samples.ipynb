## Entropy

```python
entropy={}

entropy['0']=model_instance.calculate_entropy()
entropy['10']=model_instance_10.calculate_entropy()
entropy['20']=model_instance_20.calculate_entropy()
entropy['30']=model_instance_30.calculate_entropy()
entropy['40']=model_instance_40.calculate_entropy()
entropy['50']=model_instance_50.calculate_entropy()
entropy['60']=model_instance_60.calculate_entropy()
entropy['70']=model_instance_70.calculate_entropy()
entropy['80']=model_instance_80.calculate_entropy()
entropy['90']=model_instance_90.calculate_entropy()

print(entropy)

```

```python
entropy_min=min(entropy.values())
entropy_max=max(entropy.values())
```

```python
keys=list(entropy.keys())
vals=[float(entropy[k]) for k in keys]
fig, ax=plt.subplots()
sns.barplot(x=keys, y=vals, ax=ax)
ax.set_ylim(round(entropy_min,1),round(entropy_max*1.001,2))
plt.show()
```

## Train Neural Network

```python
# Prepare dataframe to collect performance measurements
performance_df=pd.DataFrame(columns=['reduction','accuracy','precision','recall','auc_value', 'f1_score'])
performance_gt_df=pd.DataFrame(columns=['reduction','accuracy','precision','recall','auc_value', 'f1_score'])
```

```python
# Base
model_instance.build_neural_network(num_layers, num_neurons_per_layer)
model_instance.train_neural_network(num_epochs, batch_size)
accuracy, precision, recall, auc, F1_Score=model_instance.performance_measurement()
performance={'reduction':0.0,'accuracy':accuracy,'precision':precision,'recall':recall,'auc_value':auc, 'f1_score':F1_Score}
performance_df=performance_df.append(performance, ignore_index=True)

# Ground truth
accuracy, precision, recall, auc, F1_Score=model_instance.performance_measurement_ground_truth()
performance={'reduction':0.0,'accuracy':accuracy,'precision':precision,'recall':recall,'auc_value':auc, 'f1_score':F1_Score}
performance_gt_df=performance_gt_df.append(performance, ignore_index=True)
```

```python
# Data distribution
model_instance.plot_histogram(num_bins=20)
```

```python
# Build neural network
model_instance.build_neural_network(num_layers, num_neurons_per_layer)
```

```python
model_instance.train_neural_network(num_epochs=num_epochs,batch_size=batch_size)
```

```python
# Plot history
model_instance.plot_training_history()
```
